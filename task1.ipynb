```python
import random

# 生成1到240的数字列表
numbers = list(range(1, 241))

# 打乱顺序确保随机性
random.shuffle(numbers)

# 将打乱后的数字分成24组，每组10个
groups = [numbers[i*10 : (i+1)*10] for i in range(24)]

# 输出结果
for idx, group in enumerate(groups, 1):
    print(f"第{idx:2d}组: {sorted(group)}")
```
